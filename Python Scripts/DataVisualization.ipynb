```python
import pickle
import pandas as pd
```

```python
path="./../PPG_FieldStudy/S1.pkl"
```

```python
with open(path, "rb") as f:
    data = pickle.load(f, encoding="latin-1")
```

```python
data.keys()
```

```python
pd.DataFrame(data["rpeaks"])
```

```python
pd.DataFrame(data["signal"])
```

```python
pd.DataFrame(data["activity"])
```

```python
dataframes  = []
pbms = []
ok  = []
for k in data.keys():
    try:
        df = pd.DataFrame(data[k]) 
        dataframes.append(df)
        ok.append(k)
    except Exception as e:
        pbms.append(k + " : " + str(e))
print(f"clefs ok = {ok}")
print(f"clefs avec un probleme = {pbms}")
```

```python
for df in dataframes:
    print(df.shape)
```

```python
for df in dataframes:
    print(df.sample(6))
```

```python
for df in dataframes:
    print(df.columns)
```

```python
for df in dataframes:
    print(df.dtypes)
```
